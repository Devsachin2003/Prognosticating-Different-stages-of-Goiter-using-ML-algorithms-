# Decision Tree Classifier for Goiter Detection

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.impute import SimpleImputer
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from google.colab import drive
```

## 1. Data Loading
Load the dataset from the CSV file.

```python
# Mount Google Drive
#drive.mount('/content/drive')

# Load the dataset
file_path = '/content/drive/MyDrive/Copy of sick_csv.csv'
df = pd.read_csv(file_path)
```

## 2. Data Preprocessing
Convert categorical variables to numerical formats using mapping and one-hot encoding.

```python
# Replace 'f' with 0 and 't' with 1 in 'goitre' column
df['goitre'] = df['goitre'].map({'f': 0, 't': 1})

# Convert categorical columns to numerical using Label Encoding
binary_categorical_cols = ['sex', 'on_thyroxine', 'query_on_thyroxine', 'on_antithyroid_medication',
                           'sick', 'pregnant', 'thyroid_surgery', 'I131_treatment', 'query_hypothyroid',
                           'query_hyperthyroid', 'lithium', 'tumor', 'hypopituitary', 'psych',
                           'TSH_measured', 'T3_measured', 'TT4_measured', 'T4U_measured', 'FTI_measured',
                           'TBG_measured']

multi_categorical_cols = ['referral_source', 'Class']

for col in binary_categorical_cols:
    df[col] = df[col].map({'f': 0, 't': 1})

# One-hot encode multi-category columns
df = pd.get_dummies(df, columns=multi_categorical_cols)
```

## 3. Feature Selection and Imputation
Define features (X) and target (y), and handle missing values.

```python
# Define features and target variable
X = df.drop("goitre", axis=1)  # Features
y = df["goitre"]  # Target

# Impute missing values using SimpleImputer
imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X)
```

## 4. Model Training and Evaluation
Split the data into training and testing sets, train the Decision Tree Classifier, and evaluate its accuracy.

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_imputed, y, test_size=0.0019, random_state=42)

# Train the Decision Tree classifier on imputed data with limited depth
dt_classifier = DecisionTreeClassifier(max_depth=230, random_state=42)
dt_classifier.fit(X_train, y_train)
# Make predictions on the test set
y_pred = dt_classifier.predict(X_test)
accuracy_decissionTree = accuracy_score(y_test, y_pred)
# Evaluate the model
print("Accuracy:", accuracy_decissionTree )
```
